```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
se = pd.read_csv("worldse.csv")
se
```

```python
se.info()
```

```python
se.isnull().sum() #Capital gains is faulty so we're cutting it
```

```python
sen = se.drop(columns=['Capital Gains'])
sen
```

```python
sen.describe() #FIND Q1, Q3
```

```python
#make a copy before cutting to avoid irreversable damage
cleaned = sen.copy()
cols = ['Close', 'High']

for col in cols:
    Q1 = cleaned[col].quantile(0.25)
    Q3 = cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    cleaned = cleaned[
        (cleaned[col] >= Q1 - 1.5 * IQR) &
        (cleaned[col] <= Q3 + 1.5 * IQR)
    ]
cleaned = cleaned.drop_duplicates()
cleaned #Cutting outliers with maths
```

```python
#Standard Scaler because there are contrasts in numbers
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cleaned[['Open','High','Low','Close','Volume']] = scaler.fit_transform(
    cleaned[['Open','High','Low','Close','Volume']]
)
cleaned #after standard scaler
```

```python
#Making the Date more useful
cleaned['Date'] = pd.to_datetime(cleaned['Date'], utc=True)
cleaned['Year'] = cleaned['Date'].dt.year
cleaned['Month'] = cleaned['Date'].dt.month
cleaned['Day'] = cleaned['Date'].dt.day
cleaned['DayofWeek'] = cleaned['Date'].dt.dayofweek
cleaned = cleaned.sort_values('Date')
cleaned

```

```python
#Lags & Moving Averages (Show History by 1 & 5 Rows)
for col in ['Close', 'High']:
    cleaned[f'{col}_Lag1'] = cleaned[col].shift(1)
    cleaned[f'{col}_Lag5'] = cleaned[col].shift(5)
    cleaned[f'{col}_MA5'] = cleaned[col].rolling(5).mean()
    cleaned[f'{col}_MA20'] = cleaned[col].rolling(20).mean()
cleaned
```

```python
from sklearn.preprocessing import LabelEncoder

encoded = cleaned.copy()
label_cols= ['Ticker', 'Brand_Name', 'Country']
for col in label_cols:
    le = LabelEncoder()
    encoded[col] = le.fit_transform(encoded[col].astype(str))
encoded = pd.get_dummies(encoded, columns=['Industry_Tag'])
print(encoded[['Ticker', 'Brand_Name', 'Country']].head(10))
print(encoded.head(3))
```

```python
from sklearn.ensemble import RandomForestRegressor #regressor
from sklearn.multioutput import MultiOutputRegressor #cus 2 outputs close n high
from sklearn.model_selection import train_test_split 
X = encoded.drop(columns=['High','Close','Date'])
y = encoded[['High','Close',]]

print("X shape:", X.shape)
print("y shape:", y.shape)
print(X.dtypes)

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)
model = MultiOutputRegressor(RandomForestRegressor(
    n_estimators=50,   # fewer trees
    max_depth=15,      # limit depth of each tree
    random_state=42,
    n_jobs=-1          # use all CPU cores
))
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Shape of predictions:", y_pred.shape)
print("First 5 predictions:\n", y_pred[:5])
```
